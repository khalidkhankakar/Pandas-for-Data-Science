```python
import pandas as pd
```

```python
df = pd.read_csv('mtcars.csv')
df.head()
```

```python
# .loc[] vs .iloc[] 
# df.loc[2] # get row by index - label based
df.iloc[2] # position based
```

```python
df.loc[1:3]      # includes 3
df.iloc[1:3]     # excludes 3
```

```python
df.loc[1:3 , ['mpg', 'cyl']] # get rows 1 to 3 and columns mpg and cyl
```

###  Boolean Indexing (Filtering Rows with Conditions)

```python
df[df['mpg'] > 30] 
```

```python
df[(df['hp'] > 200) & (df['mpg'] < 300)] # and condition
```
