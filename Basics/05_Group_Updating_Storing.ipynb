#### Group Operations

```python
import pandas as pd
```

```python
df = pd.read_csv('mtcars.csv')
df.head()
```

```python
car = df.groupby('model')
car.first().head(3)
```

```python
car_hp = df.groupby(['model', 'hp'])
car_hp.first().head(3)
```

```python
print(df.groupby('cyl').mean(numeric_only=True))
```

Sorting Data

```python
df.sort_values('hp',ascending=False).head()
```

```python
df.sort_values(by=['hp','mpg'], ascending=[False, True]).head()
```

Updating values

```python
df['carb'] = df['carb'] * 2
df.head()
```

```python
df['vs'].replace({0: 'V', 1: 'S'}, inplace=True)    
df.head()
```

```python
df.loc[df['model'] == 'Mazda RX4', 'mpg'] = 100
df[df['model'] == 'Mazda RX4']
```
