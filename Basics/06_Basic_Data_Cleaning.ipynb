### Basic Data cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df.head()
```

#### Handle missing values

```python
print(df.isnull().sum())

# Drop the rows with missing values
df_cleaned = df.dropna()
df_cleaned.head()
```

```python
# Fill the missing values with zeros
df_filled = df.fillna(1200)
df_filled.head()
```

```python
# Fill the missing values with the mean of the column
df_fill_am = df['am'].fillna(df['am'].mean())
df_fill_am.head()
```
